```python
import numpy as np
import cv2
import glob
import os
import matplotlib.pyplot as plt
```

## pre processing:
**gray scale** and **resize** images and save them in 'Dataset' folder

```python
INPUT_FOLDER = 'Images'
j=0
```

```python
os.mkdir('Dataset')
```

```python
for img in glob.glob(INPUT_FOLDER + "/*.jpg"):
    image = cv2.imread(img)
    imgNew = cv2.resize(image ,dsize = (50,50))
    gray = cv2.cvtColor(imgNew, cv2.COLOR_BGR2GRAY)  
    cv2.imwrite(os.path.join('./Dataset','{}.jpeg'.format(j)), gray)  
    j+=1
    cv2.imshow('image' , gray)
    cv2.waitKey(30)
```

```python
cv2.destroyAllWindows()
```

```python
DATASET_PATH = "./Dataset"
IMAGE_LIST = os.listdir(DATASET_PATH)
```

#### now we must choose our image as main image

```python
CHOSEN_IMG=int( input('choose image index ("find it in dataset folder"):') )
```

```python
orginalImage = cv2.imread (os.path.join(DATASET_PATH,'{}.jpeg'.format(CHOSEN_IMG)),0)
imgplot = plt.imshow(orginalImage,cmap='gray')
```

### save images vectors in a Array:
each **row** in Vectorize_img is a reshaped image

```python
vectorize_imgs = np.zeros((len(IMAGE_LIST),50*50))
j=0
```

```python
for image_name in IMAGE_LIST:
    img = cv2.imread (os.path.join(DATASET_PATH,image_name),0)
    vectorize_imgs[j,:] = img.flatten()
    im = np.array(img,dtype=np.float32)
    j+=1
```

```python
vectorize_imgs.shape
```

### Use PCA for Eigen Faces:

```python
mean, eigenFaces = cv2.PCACompute(vectorize_imgs, mean=None, maxComponents=len(IMAGE_LIST))
print(eigenFaces.shape)
```

```python
fig=plt.figure(figsize=(10, 10))
cols=6
rows=5
k=0
for i in range (1,cols*rows+1):
    eigenFace = eigenFaces[k,:].reshape((50,50))
    fig.add_subplot(rows,cols,i)
    plt.imshow(eigenFace,cmap='gray')
    k+=1
plt.show()
k=0
```

## calculate coefficients (a):
### orginalImage =  a1 * eigenFace1 + a2 * eigenFace2 + ... + ak * eigenFacek

```python
coefficient = np.dot(eigenFaces , vectorize_imgs[CHOSEN_IMG,:])
```

now we can make our **orginal Image** with **Formula**

```python
k = int(input(" how many eigen faces must calculate? "))
```

```python
coefficient = coefficient.reshape((100,1))
finalVector = np.dot(coefficient[0:k].T , eigenFaces[0:k,:])
finalVector.shape
finalImage = finalVector.reshape((50,50))
```

### orginal Image and Eigen faced Image:

```python
fig=plt.figure()
fig.add_subplot(1,2,2)
plt.imshow(finalImage,cmap='gray')
plt.title('Eigen Face')
fig.add_subplot(1,2,1)
plt.imshow(orginalImage,cmap='gray')
plt.title('Orginal Image')
plt.show;
```
